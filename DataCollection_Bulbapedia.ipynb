### Import requests library

```
import requests

URL="https://bulbapedia.bulbagarden.net/wiki/List_of_Pok%C3%A9mon_by_National_Pok%C3%A9dex_number"
```

### Load page

```
page=requests.get(URL)
```

### Parse HTML data

```
from bs4 import BeautifulSoup

soup = BeautifulSoup(page.content, 'html.parser')
```

### Find all tables that contain Pokemon details

```
# Get main content <div>
poke_content=soup.find(id='mw-content-text')

# Get all <table> elements
poke_tables=poke_content.find_all('table')
```

### Get total number of current Pokemon generations

```
max_gens = len(poke_tables) - 4
max_gens
```

### Parse info for each generation and store in JSON files

```
info_start=3
all_gen_jsons = []
for gen_num in range(1, max_gens + 1):
  print("Generation " + str(gen_num))
  gen_json = []
  gen_list = poke_tables[gen_num]

  for i in range(info_start, len(gen_list.contents), 2):
    poke_info=gen_list.contents[i]
    kdex=poke_info.contents[1].text.strip()
    ndex=poke_info.contents[3].text.strip()
    name=poke_info.contents[7].text.strip()
    type1=poke_info.contents[9].text.strip()
    URLSoup = BeautifulSoup(str(poke_info.contents[7]))
    link = URLSoup.find('a', href=True)['href']
    link = link.replace("(", "%28").replace(")","%29")
    if len(poke_info.contents) > 10:
        type2=poke_info.contents[11].text.strip()
        gen_json.append({
            "kdex": kdex,
            "ndex": ndex,
            "name": name,
            "type1": type1,
            "type2": type2,
            "URL": "https://bulbapedia.bulbagarden.net/" + link
        })
    else:
        gen_json.append({
            "kdex": kdex,
            "ndex": ndex,
            "name": name,
            "type1": type1,
            "URL": "https://bulbapedia.bulbagarden.net/" + link
        })
  all_gen_jsons.append(gen_json)
  print(gen_json)
```

### Download each Generation JSON to computer from Google Colab

```
from google.colab import files
import json
for i in range(0, max_gens):
  filename = "gen" + str(i+1) + ".json"
  print(filename)
  gen_i = all_gen_jsons[i]
  jsonString = json.dumps(gen_i)

  with open(filename, 'w') as f:
    f.write(jsonString)
  files.download(filename)
```
